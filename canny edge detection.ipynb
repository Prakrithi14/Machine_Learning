```python

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image
img = cv2.imread('C:/Users/Prakrithi/Documents/download.jpg')  # Replace with the path to your image
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # Convert to grayscale

# Apply GaussianBlur to reduce noise
blurred = cv2.GaussianBlur(gray, (5, 5), 1.4)

# Apply Canny edge detection with varying thresholds
edges_low = cv2.Canny(blurred, 50, 150)
edges_medium = cv2.Canny(blurred, 100, 200)
edges_high = cv2.Canny(blurred, 150, 250)

# Display the results
plt.figure(figsize=(12, 12))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(edges_low, cmap='gray')
plt.title('Canny Edge Detection - Low Threshold')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(edges_medium, cmap='gray')
plt.title('Canny Edge Detection - Medium Threshold')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(edges_high, cmap='gray')
plt.title('Canny Edge Detection - High Threshold')
plt.axis('off')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread('C:/Users/Prakrithi/Documents/download.jpg')  # Replace with your image file
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur to reduce noise
blurred = cv2.GaussianBlur(gray, (5, 5), 1.4)

# Apply Canny edge detection with varying thresholds
edges_low = cv2.Canny(blurred, 50, 100)
edges_medium = cv2.Canny(blurred, 100, 200)
edges_high = cv2.Canny(blurred, 150, 250)

# Display results using matplotlib
plt.figure(figsize=(12, 6))

plt.subplot(1, 4, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original")
plt.axis("off")

plt.subplot(1, 4, 2)
plt.imshow(edges_low, cmap='gray')
plt.title("Low Threshold (50,100)")
plt.axis("off")

plt.subplot(1, 4, 3)
plt.imshow(edges_medium, cmap='gray')
plt.title("Medium (100,200)")
plt.axis("off")

plt.subplot(1, 4, 4)
plt.imshow(edges_high, cmap='gray')
plt.title("High (150,250)")
plt.axis("off")

plt.tight_layout()
plt.show()
```
